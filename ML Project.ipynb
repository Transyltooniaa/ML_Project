## Distribution of Numerical Features


Visualize the distribution of numerical columns with histograms and density plots.

```python
numerical_columns = df.select_dtypes(include=[np.number]).columns
numerical_columns = [col for col in numerical_columns if col != 'Default']

# Define subplot layout
n_cols = 2
n_rows = (len(numerical_columns) + 1) // n_cols  

# Set up the figure for subplots
plt.figure(figsize=(15, 4 * n_rows))
plt.suptitle("Numerical Features Distribution", fontsize=16)

# Loop through each numerical column and create a subplot
for idx, col in enumerate(numerical_columns, 1):
    plt.subplot(n_rows, n_cols, idx)
    sns.histplot(df[col], kde=True)
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)

# Adjust layout
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

## Outliers Detection using Boxplots

Use box plots to detect any outliers in numerical columns.

## Validity Checks for Numerical Columns


Check if values in numerical columns are within expected ranges.

```python
validity_checks = {
    'Age': (0, 100),
    'Income': (0, 500000),
    'LoanAmount': (0, 1000000),
    'CreditScore': (300, 850),
    'MonthsEmployed': (0, 480),
    'InterestRate': (0, 100),
    'LoanTerm': (0, 360),
    'DTIRatio': (0, 1)
}

print("\nValidity Checks for Numerical Columns:")
for col, (min_val, max_val) in validity_checks.items():
    if col in df.columns:
        invalid_data = df[(df[col] < min_val) | (df[col] > max_val)]
        print(f"\n{col} - Invalid values found:")
        print(invalid_data[[col]] if not invalid_data.empty else "None")
```

# Data Preprocessing

## Label Encoding Training data and Testing data

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
obj_col = ['HasCoSigner','LoanPurpose','HasDependents', 'HasMortgage','MaritalStatus', 'EmploymentType', 'Education']
```

```python
for col in obj_col:
    df[col] = le.fit_transform(df[col])

df = df.drop(['LoanID'], axis=1)
```

```python
for col in obj_col:
    df_test[col] = le.fit_transform(df_test[col])

df_test = df_test.drop(['LoanID'], axis=1)
```

## Splitting training data into dependent and independent variables

```python
X = df.drop(['Default'], axis=1)
y = df['Default']
```

# Model Training

## RandomForest

```python
import seaborn as sns
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.ensemble import RandomForestClassifier

sample_submission = pd.read_csv('sample_submission.csv')

rf_param_grid = {
    'n_estimators': [100,],
    'max_depth': [10, 20, None],
    'min_samples_split': [2],
    'min_samples_leaf': [1,],
    'random_state': [42]
}

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Initialize RandomForestClassifier
rf = RandomForestClassifier()

rf_random_search = RandomizedSearchCV(
    rf, rf_param_grid, n_iter=1, cv=2, random_state=42, n_jobs=-1, scoring='accuracy'
)

rf_random_search.fit(X_train, y_train)
best_rf_model = rf_random_search.best_estimator_

y_pred = best_rf_model.predict(X_test)
```

### Model Evaluations

```python
train_accuracy = accuracy_score(y_train, best_rf_model.predict(X_train))
print("Training Accuracy:", train_accuracy)

train_class_report = classification_report(y_train, best_rf_model.predict(X_train))
print("Classification Report:\n", train_class_report)
```

### Confusion Matrix

```python
cm = confusion_matrix(y_test, y_pred)

# Plot Confusion Matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=['Class 0', 'Class 1'], yticklabels=['Class 0', 'Class 1'])
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.title('Confusion Matrix')
plt.show()
```

### Saving the Predictions

```python
df_test_pred = best_rf_model.predict(df_test)

submission_df = sample_submission.copy()
submission_df['Default'] = df_test_pred  

submission_df.to_csv('RandomForest_predictions.csv', index=False)
print("Predictions saved to 'RandomForest_predictions.csv'")
```

## XGBoost 

```python
from xgboost import XGBClassifier
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
import seaborn as sns

sample_submission = pd.read_csv('sample_submission.csv')

xgb_param_grid = {
    'n_estimators': [100, 200,],
    'max_depth': [3, 4],
    'learning_rate': [0.01, 0.05],
    'subsample': [0.7, 0.8,],
    'colsample_bytree': [0.7, 0.8],
    'eval_metric': ['logloss'],  
    'random_state': [42]
}

xgb = XGBClassifier()

random_search = RandomizedSearchCV(
    xgb, param_distributions=xgb_param_grid, n_iter=2, cv=2, random_state=42, n_jobs=-1, scoring='roc_auc'
)

random_search.fit(X, y)

xgb_best = random_search.best_estimator_

# Predict on df_test (assuming df_test is already defined and preprocessed)
df_test_pred = xgb_best.predict(df_test)

# Evaluate on the training set 
train_pred = xgb_best.predict(X)
train_accuracy = accuracy_score(y, train_pred)
train_class_report = classification_report(y, train_pred)
```
